## RL 交互回路（一张图）

```mermaid
flowchart LR
  A[Agent\n策略 π(a|s)] -- 选择动作 a --> E[Environment\n环境]
  E -- 返回状态 s' 与奖励 r --> A
  A -. 用 (s,a,r,s') 学习更新 .-> A
```

## 核心：试错 + 长期回报
- 状态 s：处境
- 动作 a：选择
- 奖励 r：反馈
- 策略 π：怎么选
- 折扣 γ：多看重未来
